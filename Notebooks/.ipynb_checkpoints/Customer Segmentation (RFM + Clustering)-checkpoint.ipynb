```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("merged_orders_customers.csv")
df.head()
```

```python
df.columns
```

```python
date_cols = ['order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Keep only delivered orders
df = df[df['order_status'] == 'delivered']
```

```python
(df['customer_unique_id_x'] == df['customer_unique_id_y']).value_counts()
```

```python
df['customer_unique_id'] = df['customer_unique_id_x']
```

```python
df.drop(columns=['customer_unique_id_x','customer_unique_id_y'], inplace=True)
```

```python
snapshot_date = df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

rfm = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'nunique',
    'total_payment_value': 'sum'
})

rfm.rename(columns={
    'order_purchase_timestamp': 'Recency',
    'order_id': 'Frequency',
    'total_payment_value': 'Monetary'
}, inplace=True)

rfm.head()
```

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

```python
sse = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    sse.append(kmeans.inertia_)

# Plot Elbow curve
plt.figure(figsize=(8,4))
plt.plot(K_range, sse, marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("SSE")
plt.title("Elbow Method for Optimal K")
plt.show()

# Based on elbow, choose K (example: 4)
k_opt = 4
kmeans = KMeans(n_clusters=k_opt, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_labels = {
    0: 'Loyal',
    1: 'Potential Loyalists',
    2: 'Regular',
    3: 'At Risk'
}
rfm['Segment'] = rfm['Cluster'].map(cluster_labels)

# Check cluster stats
print(rfm.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(2))
```

```python
# Bar chart â†’ Customer count per segment
plt.figure(figsize=(6,4))
sns.countplot(x='Segment', data=rfm, order=['Loyal','Potential Loyalists','Regular','At Risk'])
plt.title("Customer Count per Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Scatter plots â†’ Recency vs Monetary
plt.figure(figsize=(6,4))
sns.scatterplot(x='Recency', y='Monetary', hue='Segment', data=rfm)
plt.title("Recency vs Monetary")
plt.show()
```

```python
# Scatter plots â†’ Frequency vs Monetary
plt.figure(figsize=(6,4))
sns.scatterplot(x='Frequency', y='Monetary', hue='Segment', data=rfm)
plt.title("Frequency vs Monetary")
plt.show()
```

```python
rfm.to_csv("rfm_customer_segmentation.csv", index=True)
```
